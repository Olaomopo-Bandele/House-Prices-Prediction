MODELLING

```python
#spliting my test and train set into 70/30
X = miami_c.drop('SALE_PRC', axis=1)
y= miami_c['SALE_PRC']
```

```python
from sklearn.preprocessing import StandardScaler    #feature scaling
scaler = StandardScaler() 
Xscaled = scaler.fit_transform(X)
```

```python
from sklearn import metrics
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( Xscaled, y, test_size=0.3, random_state=101)
```

```python

regressor =SVR()
rfr = RandomForestRegressor(n_estimators=60)
```

```python
def fitandprint(model, X_train, X_test, y_train, y_test):  # take the model, train data and test data as input
    
    model.fit(X_train, y_train)   # fit the model using the train data

    prediction = model.predict(X_test)     # model predictions on the test data
    
    mse = metrics.mean_squared_error(y_test, prediction) #calculate mse on the test data

    r2 = metrics.r2_score(y_test, prediction)  # calculate the r squared value on the test data
    
    rmse = np.sqrt(metrics.mean_squared_error(y_test, prediction))   # Root mean squared error
  
    return r2, rmse, mse, prediction
```

```python
result = {} #A dictionary to store output of the metrix
```

```python
#for loop to loop through each of the models and append its output to models
for model, name  in zip([ regressor,rfr], [ 'SVR', 'Random Forest']):
    result[name] = fitandprint(model, X_train, X_test, y_train, y_test)
```

```python
result1 = pd.DataFrame(np.array(list(result.values()))[:,:-1],    # make a dataframe out of the metrics from result dictionary 
                       columns= ['R Squared', 'RMSE', 'MSE'],
                      index= result.keys())   # use the model names as index

result1.index.name = 'Model'   # name the index of the result1 dataframe as 'Model'

result1 
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split( Xscaled, y, test_size=0.3, random_state=101)

rfr = RandomForestRegressor(n_estimators=60)
```

```python
rfr.fit(Xtrain, ytrain)
```

```python
predictions = rfr.predict(Xtest)
df = pd.DataFrame({'Actual': ytest, 'Predicted': predictions})
df
```

```python
test = pd.DataFrame({'Predicted value':predictions, 'Actual value':ytest})
fig= plt.figure(figsize=(16,8))
test = test.reset_index()
test = test.drop(['index'],axis=1)
plt.plot(test[:50])
plt.legend(['Actual value','Predicted value'])
plt.title("Train and test for Random forest regressor")
```
